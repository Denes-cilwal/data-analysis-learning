```python
# shallow Copy vs Deep Copy in Arrays
# Shallow copy creates a new object but shares the same data. Deep copy creates a completely independent copy.
```

```python
 # No Copy (Just a Reference)
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
b = arr            # no copy, just another name

b[0] = 99
print(arr)         # [99, 2, 3, 4, 5] ← original changed!
print(b is arr)    # True (same object)
```

```python
 # Shallow Copy (View)
arr = np.array([1, 2, 3, 4, 5])
b = arr.view()     # shallow copy

b[0] = 99
print(arr)         # [99, 2, 3, 4, 5] ← original changed!
print(b is arr)    # False (different object, but shared data)
```

```python
arr = np.array([1, 2, 3, 4, 5])
b = arr.copy()     # deep copy

b[0] = 99
print(arr)         # [1, 2, 3, 4, 5] ← original unchanged!
print(b is arr)    # False
```

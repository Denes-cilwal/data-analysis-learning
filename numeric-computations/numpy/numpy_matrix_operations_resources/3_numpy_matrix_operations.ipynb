```python
import numpy as np
```

```python
# Sales data for Quarter 1 (Matrix 1)
# Rows represent different products, columns represent different regions

q1 = np.array([
    [200, 220, 250],  # Product A
    [150, 180, 210],  # Product B
    [300, 330, 360]   # Product C
])

q2 = np.array([
    [209, 231, 259],  # Product A
    [155, 192, 222],  # Product B
    [310, 340, 375]   # Product C
])

```

#### Addition, Subtraction, Multiplication, Division

```python
# combined sales of Q1 and Q2
q1 + q2 # no need to apply  loop here since this is vector operations
```

```python
# sales units growth in q2 compared to q1
# percent growth in quarter 2
q2 - q1
```

```python
# % sales units growth in q2 compared to q1
# % Growth = (New − Old) / Old × 100
(q2 - q1) / q1 * 100
```

```python
# ROWS IS PRODUCTS AND COLUMN IS REGION, USA, CANADA, CHINA
"""
Units sold = 100
Price per unit = 100
Revenue = 100 × 100 = 10,000

Revenue (also called Gross Revenue)

What it is:
The total money earned from sales before anything is deducted.

Includes:
Sales value
Before discounts
Before returns
Before taxes, fees, commissions
"""
prices = np.array([
    [10, 12, 11],  # Prices for Product A in different regions
    [8,  9,  10],  # Prices for Product B
    [15, 16, 17]   # Prices for Product C
])

# prices of each product - for revenue (q1 unit of sales * prices)
q1_revenue = q1 * prices
q1_revenue
```

```python
# flat discount given 20 %
# Multiplying by a percentage means “take that fraction”, not “increase”
"""
Multiplying by a number less than 1 → value decreases
Multiplying by a number greater than 1 → value increases
10,000 × 2   = 20,000  (increase)
10,000 × 1  = 10,000  (same)
10,000 × 0.5 = 5,000  (decrease)
10,000 × 0.02 = 200   (big decrease)

"""

q1_discounts = q1_revenue*0.02
q1_discounts
```

```python
 # Revenue and Net Revenue are related but not the same.
"""
Net Revenue
What it is:
The money the company actually keeps from sales after deductions.
Revenue minus:
Discounts
Returns / refunds
Allowances
Sales commissions
(sometimes) indirect taxes

 """
net_revenue = q1_revenue - q1_discounts
net_revenue
```

```python
np.sum(q1_discounts)
```

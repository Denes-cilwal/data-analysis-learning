```python
# why python occupies more memory ?
# python list array is heterogeneous. if you have python array one element can be string, another can be float and soon
# to support this it needs to store meta-data in the memory and the address are not contiguous, they might be scattered in the memory
# you can see the memory is fragmented so each of the byte store meta-data because of that total size occupied by python list is higher
# since memory is fragmented accessing element is also slower takes time while data is stored in contiguous memory location
```

```python
# measuring performance in list

import time
size = 10
l1 = list(range(size))
l2 = list(range(size))
start = time.time()
l3 = [x+y for x, y in zip(l1, l2)]
print(l3)
end = time.time()

print("Python list took",end-start)
```

```python
n1 = np.arange(size)
n2 = np.arange(size)

start = time.time()
n3 =  n1 + n2
end = time.time()

print("NumPy took",end-start)

# took less time because we do not need to do iteration (loops) instead we use Vectorized operations that is programming technique that performs a single operation on entire arrays (vectors/matrices) at once, rather than looping through each element,
```

```python
# sorting use case in numpy
a = np.sort([3,2,1])
print(a)

b = np.sort([3,2,1])
print(b)

# sort 2D by rows
n =  np.array([[3,1,2],[9,7,8]])

"""
Think of it as 2 rows √ó 3 columns:
Row 0: [3, 1, 2]
Row 1: [9, 7, 8
"""


print(np.sort(n, axis=1))# sort each row
"""
axis=1 = sort each row
Row means ‚Äúhorizontal line‚Äù.
Row0 is: [3, 1, 2] ‚Üí sorted ‚Üí [1, 2, 3]
Row1 is: [9, 7, 8] ‚Üí sorted ‚Üí [7, 8, 9]
"""

"""
axis=0 = sort each column
Column means ‚Äúvertical line‚Äù.
Column0 is: [3, 9] ‚Üí sorted ‚Üí [3, 9]
Column1 is: [1, 7] ‚Üí sorted ‚Üí [1, 7]
Column2 is: [2, 8] ‚Üí sorted ‚Üí [2, 8]

Why did axis=0 not change anything here?

Because in each column, the top number is already smaller than the bottom number:
3 < 9
1 < 7
2 < 8
So sorting columns doesn‚Äôt move anything.
"""


"""
For your array, axis=2 is not valid.

Why?

Your array:

m = np.array([[3, 1, 2],
              [9, 7, 8]])

has 2 dimensions:
m.ndim   # 2
m.shape  # (2, 3)


So the valid axes are:

axis = 0 ‚Üí rows (vertical)
axis = 1 ‚Üí columns (horizontal)
There is no axis 2.
AxisError: axis 2 is out of bounds for array of dimension 2
‚Üì axis=0 (down rows)

[ 3  1  2 ]  ‚Üí axis=1 (across columns)
[ 9  7  8 ]

What each axis really means
üîπ axis = 0 ‚Üí move down (vertical)

Operation happens column by column
Rows are compared within each column

üîπ axis = 1 ‚Üí move across (horizontal)

Operation happens row by row
Columns are compared within each row
"""
```
